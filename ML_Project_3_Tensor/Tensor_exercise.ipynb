```python
#Scalar, 0-rank tensor/0-dimensional tensor
import numpy as np
x =np.array(5)
print("The value of x:",x)
print("The rank of x:",x.ndim)
print("The data type of x:",x.dtype)
print("The shape of x:",x.shape)
```

```python
#Vector, 1-rank tensor/1-dimensional tensor
y = np.array([1,2,3,4,5]) #This is a 1-dimensional tensor with 5 elements.
print("The value of y:",y)
print("The rank of y:",y.ndim)
print("The data type of y:",y.dtype)
print("The shape of y:",y.shape) #It is 5-dimensional vector.

#Vector dataset in ML
from tensorflow.keras.datasets import boston_housing
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
print("The shape of x_train:",x_train.shape) #It is 404-dimensional vector.
print("The shape of the first sample of x_train:",x_train[0].shape) #It is 13-dimensional vector.
print("The shape of y_train:",y_train.shape) #It is 404-dimensional vector.

# Dot product of two vectors
a = np.array([1,-1.8,1,1,2])
b = np.array([1,6,7,8,9])
output = np.dot(a,b) #This is the dot product of two vectors a and b.

print("The dot product of a and b:",output)

```

```python
#Matrix, 2-rank tensor/2-dimensional tensor
# Matrix is a 2-dimensional tensor,the shape is (samples, features), where samples is the number of samples in the dataset, and features is the number of features in each sample.
print("the contents of x_train:",x_train)
print("the shape of x_train:",x_train.shape)
```

```python
#Create 3-rank tensor/3-dimensional tensor
c = np.array([[[1,22,4,78,2],[2,59,6,56,1],[3,31,8,54,0]],[[4,56,9,34,1],[5,78,8,35,2],[6,34,7,36,0]],[[7,12,6,37,1],[8,45,5,38,2],[9,23,4,39,0]]])
print("the contents of c:",c)
# In 2-dimensional tensor, the shape is samples and features, but in 3-dimensional tensor, the shape is samples, timesteps and features. 
# In this case, the shape of c is (3, 3, 5), where 3 is the number of samples, 3 is the number of timesteps, and 5 is the number of features.
print("the shape of c:",c.shape)
```
